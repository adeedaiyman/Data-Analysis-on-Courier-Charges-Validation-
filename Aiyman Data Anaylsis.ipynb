```python
# importing required libraries
import pandas as pd
import numpy as np
```

```python
#Importing all the xlsx files
order_df = pd.read_excel('/home/bashar/Downloads/ML/Assignment/Company X - Order Report.xlsx')
pincode_zones_df = pd.read_excel('/home/bashar/Downloads/ML/Assignment/Company X - Pincode Zones.xlsx')
sku_master_df = pd.read_excel('/home/bashar/Downloads/ML/Assignment/Company X - SKU Master.xlsx')
invoice_df = pd.read_excel('/home/bashar/Downloads/ML/Assignment/Courier Company - Invoice.xlsx')
rate_card_df = pd.read_excel('/home/bashar/Downloads/ML/Assignment/Courier Company - Rates.xlsx')
```

```python
order_df.head()
```

```python
#renaming the column from 'ExternOrderNo' to 'Order ID'
order_df.rename(columns={'ExternOrderNo': 'Order ID'}, inplace=True)
order_df.head()
```

```python
pincode_zones_df.head()                 
```

```python
#extracting the column name 'Delivery Zones as Per X' from column 'Zone'
pincode_zones_df['Delivery Zones as Per X'] = pincode_zones_df['Zone']
```

```python
#creating a function to extract the column name 'Weight slab as per X (KG)' with the help of column name 'Delivery Zones as Per X'
def calculate_weight_slab(Zone):
    if Zone == 'a':
        return 0.25
    elif Zone == 'b':
        return 0.5
    elif Zone == 'c':
        return 0.75
    elif Zone == 'd':
        return 1.25
    elif Zone == 'e':
        return 1.5
    else:
        return None  # Handle other cases if needed

pincode_zones_df['Weight slab as per X (KG)'] = pincode_zones_df['Delivery Zones as Per X'].apply(calculate_weight_slab)
```

```python
pincode_zones_df.head()
```

```python
sku_master_df.head()     
```

```python
#extracting 'totsl weight as per x'with the help of 'weight (g)' column
sku_master_df['Total Weight as per X (KG)'] = sku_master_df['Weight (g)'] * order_df['Order Qty'] / 1000
column_to_drop = ['Weight (g)']
sku_master_df.drop(columns=column_to_drop, inplace=True)
sku_master_df.head()
```

```python
invoice_df.head()
```

```python
#renaming the column 'charged weight' to 'weight slab charged by courier company'
invoice_df.rename(columns={'Charged Weight': 'Weight Slab Charged by Courier Company (KG)'}, inplace=True)
```

```python
#in the invoice dataset creating a column name 'delivery zones charged by the courier company' with the help of 'zone'
invoice_df['Delivery Zones Charged by Courier Company'] = pincode_zones_df['Zone']
```

```python
#creating a function name 'calculate_weight' to extract a 'total weight as per the courier company'
def calculate_weight(Zone):
    if Zone == 'a':
        return 0.25
    elif Zone == 'b':
        return 0.5
    elif Zone == 'c':
        return 0.75
    elif Zone == 'd':
        return 1.25
    elif Zone == 'e':
        return 1.5
    else:
        return None  # Handle other cases if needed

invoice_df['Total weight as per Courier Company (KG)'] = invoice_df['Delivery Zones Charged by Courier Company'].apply(calculate_weight)
```

```python
invoice_df.head()
```

```python
rate_card_df.head()
```

```python
#creating a dictionary to replace the values of a speicific column
conditions =[
    (rate_card_df['Zone'] =='A'),
    (rate_card_df['Zone'] =='B'),
    (rate_card_df['Zone'] == 'C'),
    (rate_card_df['Zone'] == 'D'),
    (rate_card_df['Zone'] == 'E')
] 

choices = ['a','b','c','d','e']
rate_card_df['Zone'] = np.select(conditions, choices)

rate_card_df
```

```python
# merging the order_df and sku_df based on the common identifier which is 'SKU'
merged_df = order_df.merge(sku_master_df, on = 'SKU')
merged_df.head()
```

```python
# merging the 'merged_df' and 'invoice_df based' on the common identifier which is 'Order ID'
merged_df = merged_df.merge(invoice_df, on = 'Order ID')
merged_df.head()
```

```python
# merging the 'merged_Df' and 'pincode_zones_df' based on the common identifier which is 'Warehouse Pincode'
merged_df = merged_df.merge(pincode_zones_df, left_on='Warehouse Pincode', right_on='Warehouse Pincode', suffixes=('', '_pincode_Zone'))
merged_df.head()
```

```python
# merging the 'merged_df' and 'rate_Card_df' based on the common identifier which is 'Zone'
merged_df = merged_df.merge(rate_card_df, on = 'Zone')
merged_df.head()
```

```python
#creating a function name 'calculate_forward_carge()' to extract a required details and the column name 'forward_charge'
def calculate_forward_charge(row):
    # Access the necessary columns from the row
    weight = row['Weight Slab Charged by Courier Company (KG)']
    zone = row['Delivery Zones Charged by Courier Company']
    weight_slabs = row['Weight Slabs']
    forward_fixed_charge = row['Forward Fixed Charge']
    forward_additional_slab_charge = row['Forward Additional Weight Slab Charge']

    # Implement forward charge calculation logic here based on weight, zone, and rate card
    # Calculate the forward charge based on the weight and rate card
    forward_charge = forward_fixed_charge

    # Calculate the additional slab charges if applicable
    if weight > weight_slabs:
        additional_weight = weight - weight_slabs
        additional_slabs = additional_weight / forward_additional_slab_charge
        forward_charge += (additional_slabs * forward_additional_slab_charge)

    return forward_charge

# Apply the calculate_forward_charge function to each row in the merged DataFrame
merged_df['Forward Charge'] = merged_df.apply(calculate_forward_charge, axis=1)
```

```python
merged_df.head()
```

```python
#creating a function name 'calculate_cod_carge()' to extract a required details and the column name 'cod_charge'
def calculate_cod_charge(row):
    # Access the necessary columns from the row
    payment_mode = row['Payment Mode']
    order_amount = row['Billing Amount (Rs.)']
    
    # Implement COD charge calculation logic here based on payment mode and order amount
    if payment_mode == 'COD':
        if order_amount <= 300:
            cod_charge = 15
        else:
            cod_charge = 0.05 * order_amount
    else:
        cod_charge = 0
    
    return cod_charge

# Apply the calculate_cod_charge function to each row in the merged DataFrame
merged_df['COD Charge'] = merged_df.apply(calculate_cod_charge, axis=1)
```

```python
merged_df.head()
```

```python
#adding three columns named 'forward charge', 'rto fixed charge' and 'cod charges' to know the 'total charges'
merged_df['Total Charges'] = merged_df['Forward Charge'] + merged_df['RTO Fixed Charge'] + merged_df['COD Charge']
merged_df.head()
```

```python
#adding three columns named 'forward charge', 'rto fixed charge' and 'cod charges' to calculating the 'Expected Charge as per X (Rs.)'
merged_df['Expected Charge as per X (Rs.)'] = merged_df['Forward Charge'] + merged_df['RTO Fixed Charge'] + merged_df['COD Charge']
merged_df.head()
```

```python
# Assuming 'Billing Amount (Rs.)' is the column containing billed charges in the 'invoice_df'
merged_df['Charges Billed by Courier Company (Rs.)'] = invoice_df['Billing Amount (Rs.)']
merged_df.head()
```

```python
merged_df['Difference Between Expected Charges and Billed Charges (Rs.)'] = merged_df['Expected Charge as per X (Rs.)'] - merged_df['Charges Billed by Courier Company (Rs.)']
merged_df.head()
```

```python
merged_df.shape
```

```python
#removing the duplicate rows
merged_df = merged_df.drop_duplicates(keep='last')
```

```python
merged_df.shape
```

```python
merged_df.fillna(0, inplace=True)
```

```python
# Calculate total orders and total invoice amount
total_orders = len(merged_df)
total_invoice_amount = merged_df['Billing Amount (Rs.)'].sum()

# Count correctly charged orders and total invoice amount
correctly_charged_orders = merged_df[
    merged_df['Expected Charge as per X (Rs.)'] == merged_df['Billing Amount (Rs.)']
]
count_correctly_charged = len(correctly_charged_orders)
total_correctly_charged_amount = correctly_charged_orders['Billing Amount (Rs.)'].sum()

# Count overcharged orders and total overcharging amount
overcharged_orders = merged_df[
    merged_df['Expected Charge as per X (Rs.)'] > merged_df['Charges Billed by Courier Company (Rs.)']
]
count_overcharged = len(overcharged_orders)
total_overcharging_amount = (overcharged_orders['Expected Charge as per X (Rs.)'] - overcharged_orders['Charges Billed by Courier Company (Rs.)']).sum()

# Count undercharged orders and total undercharging amount
undercharged_orders = merged_df[
    merged_df['Expected Charge as per X (Rs.)'] < merged_df['Charges Billed by Courier Company (Rs.)']
]
count_undercharged = len(undercharged_orders)
total_undercharging_amount = (undercharged_orders['Charges Billed by Courier Company (Rs.)'] - undercharged_orders['Expected Charge as per X (Rs.)']).sum()

# Print the results
print("Total Orders:", total_orders)
print("Total Invoice Amount:", total_invoice_amount)
print("Total Correctly Charged Orders:", count_correctly_charged)
print("Total Correctly Charged Amount:", total_correctly_charged_amount)
print("Total Overcharged Orders:", count_overcharged)
print("Total Overcharging Amount:", total_overcharging_amount)
print("Total Undercharged Orders:", count_undercharged)
print("Total Undercharging Amount:", total_undercharging_amount)
```
